```python
# Read data
row_list = []

with open('initial_data') as sanjay:
    line = sanjay.readline()
    while line: 
        row_list.append(line.split())
        line = sanjay.readline()

# Extract data from lists with 7 elements
prefix = [row_list[1][1], row_list[2][1], row_list[2][2]]
col_label = row_list[0]
del row_list[1][1], row_list[2][2], row_list[0]

# Create data frame and broadcast addittional columns
df = pd.DataFrame(row_list)
df['path2'] = df.iloc[:, 2] + ' ' + df.iloc[:, 3] + ' ' + df.iloc[:, 4]
df.iloc[1, 6] = prefix[-1] + ' ' + df.iloc[1, 6]
df['Network'] = prefix[0]

# Create final data frame
df_clean = df.iloc[:, [0, 1, 5, 6, 7]]
df_clean.columns = ['Valid', 'Next-Hop', 'Protocol', 'Path', 'Network']
df_clean = df_clean[col_label]
```
